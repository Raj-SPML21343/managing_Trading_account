```python
import csv


def calculate_total_pnl(path):
  # read csv file with csv reader
  with open(path,'r') as f:
    reader = csv.reader(f)
    next(reader)
    trades_20 = list(reader)

  def modify(l1,l2,p):
    # print(l1,l2)
    if len(l1) == 0:
      l2[-1] = str(-float(l2[-1]))
      l1.append(l2)
      return
    elif float(l1[0][-1]) > float(l2[-1]):
      l1[0][-1] = str(-float(l2[-1]) + float(l1[0][-1]))
      p += float(l2[-2])*float(l2[-1]) - float(l2[-1])* float(l1[0][-2])
      x = [l1[0][0],l2[0],l2[1],min(float(l2[-1]),float(l1[0][-1])), round(p,2), l1[0][2], l2[2],l1[0][-2],l2[-2]]
      return x
    elif float(l1[0][-1]) == float(l2[-1]):
      p += float(l2[-2])*float(l2[-1]) - float(l1[0][-1])* float(l1[0][-2])
      x = [l1[0][0],l2[0],l2[1],min(float(l2[-1]),float(l1[0][-1])), round(p,2), l1[0][2], l2[2],l1[0][-2],l2[-2]]
      l1.pop(0)
      return x
    else:
      p += float(l1[0][-1])*float(l2[-2]) - float(l1[0][-1])* float(l1[0][-2])
      l2[-1] = str(float(l2[-1])-float(l1[0][-1]))
      l1.pop(0)
      return modify(l1,l2,p)

  dict1 = {}
  total = 0
  count = 0
  for element in trades_20:
    count += 1
    if element[1] in dict1.keys():
      if element[2] == 'B':
        dict1[element[1]]+=[element]
      elif element[2] == 'S':
        x = modify(dict1[element[1]],element, 0)
        if x:
          print(x)
          total += float(x[4])
    else:
      dict1[element[1]] = [element]


  print(total)

```

```python
calculate_total_pnl(input("Please Enter your path of csv file here: "))
```
